

## WHERE
La cláusula WHERE se usa como filtro, para extraer solamente aquellos registros que cumplan una condición.

```SQL
SELECT column_name,column_name FROM table_name WHERE column_name operator value

```

````python
query = '''SELECT *
           FROM winter
           WHERE Country = 'ESP' '''

pysqldf(query)
````

````python
query = '''SELECT *
           FROM dictionary
           WHERE Code = 'FRA' '''

pysqldf(query)
````

El operador NOT invierte el valor de la condición a la que se antepone.



## LIKE

El operador LIKE se usa en una cláusula WHERE para buscar un patrón específico en una columna. 

Hay dos comodines (WILDCARDS) utilizados en conjunción con el operador LIKE: 
* % : El signo de porcentaje representa cero, uno o varios caracteres 
* _  : El subrayado representa un único carácter

```SQL
SELECT column1, column2, ... FROM table_name WHERE columnN LIKE pattern

```



## IN

El operador IN permite especificar varios valores en una cláusula WHERE.​

El operador in es un atajo para múltiples condiciones.​

```SQL 
SELECT column_name(s) FROM table_name WHERE column_name IN (value1, value2, ...)

SELECT column_name(s) FROM table_name WHERE column_name IN (SELECT STATEMENT)
```

````python
query = '''SELECT *
           FROM winter
           WHERE Year IN (2002, 1924) '''

pysqldf(query)
````



## BETWEEN

El operador BETWEEN selecciona los valores dentro de un rango determinado.


Los valores pueden ser números, texto o fechas.


El operador BETWEEN es incluyente para los valores numéricos: los valores de inicio y cierre están incluidos. Pero no lo es con los textos, en los que solo incluye el valor inicial.

```SQL
SELECT column_name(s) FROM table_name WHERE column_name BETWEEN value1 AND value2
```

````python
query = '''SELECT *
           FROM winter
           WHERE Year BETWEEN 2000 AND 2010 '''

pysqldf(query)
````



## ORDER BY

La cláusula ORDER BY se utiliza para ordenar el conjunto de resultados en orden ascendente o descendente.

ORDER BY ordena los registros en orden ascendente de forma predeterminada.

Para ordenar los registros en orden descendente, utilice la palabra clave DESC.

```SQL 
SELECT column1, column2, ... FROM table_name ORDER BY column1, column2, ... ASC|DESC
```

````python
query = '''SELECT *
           FROM dictionary
           ORDER BY GDP ASC'''

pysqldf(query)
````

````python
query = '''SELECT *
           FROM dictionary
           ORDER BY GDP DESC'''

pysqldf(query)
````
